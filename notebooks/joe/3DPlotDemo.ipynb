```python
%matplotlib widget
import numpy as np
import matplotlib.pyplot as plt

# Create a simple 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Sample 3D data (e.g., a spiral path)
t = np.linspace(0, 10, 1000)
x = np.sin(t)
y = np.cos(t)
z = t

ax.plot(x, y, z)

plt.show()
```
